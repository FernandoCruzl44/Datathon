```python
import pandas as pd
from datetime import datetime
import os
```

```python
Sales = pd.read_csv("files/salesClean.csv")
Flights = pd.read_csv("files/limpios2023.csv")
```

```python
# Preprocess the data
Flights['STD'] = pd.to_datetime(Flights['STD'])  # Convert flight_date to datetime
Flights['STA'] = pd.to_datetime(Flights['STA'])  # Convert booking_date to datetime
```

```python
#Escogemos la venta de perecederos
predSales =  Sales[Sales["ProductType"] ==  'Perecederos'] 
predSales.drop(columns=['ProductName', 'TotalSales','ProductType'], inplace=True)
print(len(predSales['Flight_ID'].unique()))

# predSales.sort_values

predSales.sort_values(by='Flight_ID')
print(predSales)
predSales = predSales.groupby('Flight_ID').sum()['Quantity']
```

```python
duplicate_flights = Flights[Flights['Flight_ID'].duplicated()]['Flight_ID']
duplicate_list = duplicate_flights.tolist()
print(len(duplicate_list))
# Filtrar DataFrame excluyendo filas con IDs en la lista
noDuplicateFlights = Flights[~Flights['Flight_ID'].isin(duplicate_list)]
print(len(Flights) - 2 * len(duplicate_list))
print(len(noDuplicateFlights))
```

```python
join = Flights.merge(predSales, on="Flight_ID",how="inner")
print(len(join))
a = join['Flight_ID'].unique()
f_unique = Flights['Flight_ID'].unique()
join.sort_values(by='Quantity', ascending=False)
```

```python
duplicate_flights = Flights[Flights['Flight_ID'].duplicated()]['Flight_ID']
duplicate_list = duplicate_flights.tolist()
print(len(duplicate_list))
# Filtrar DataFrame excluyendo filas con IDs en la lista
noDuplicateFlights = Flights[~Flights['Flight_ID'].isin(duplicate_list)]
print(len(Flights) - 2 * len(duplicate_list))
print(len(noDuplicateFlights))
```

```python
f2 = join
f2['Ocupacy'] = (f2['Passengers'] / f2['Capacity'])
f2['Month'] = f2['STD'].apply(lambda x: x.month)
f2['Time'] = (f2['STA'] - f2['STD']).dt.total_seconds() / 60
f2['Day'] = f2['STD'].apply(lambda x: x.day_of_week)
f2['Hour'] = f2['STD'].apply(lambda x: x.hour)
```

```python
f2.to_csv("files/perecederos.csv", index=False)

# Verificamos la existencia del archivo
if os.path.exists("files/perecederos.csv"):
    print("Archivo perecederos.csv creado con éxito!")
else:
    print("Error al crear el archivo files/perecederos.csv")
```
